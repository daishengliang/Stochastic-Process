## PART 1: Estimation of infection rate of virus

Recall the Reed-Frost model for spread of infection from Lab 6. At $t=0$, each individual is either susceptible or infected. The evolution over one time step is the following.   A susceptible node has a chance $\beta$ to become infected by each of its infected neighbors, with the chances from different neighbors being independent. Thus if a susceptible node has $k$ infected neighbors at time $t,$ the probability the node is *not* infected at time $t+1$ (i.e. it remains susceptible) is $(1-\beta)^k,$  and the probability the node is infected at time $t+1$ is
$1-(1-\alpha)^k.$   It is assumed that a node is removed one time step after being infected, and once a node is removed, it remains removed forever. The following code is for the simple case that a virus can spread from any node to any other node.   This corresponds to a tightly clustered population; the graph is the complete graph.   For this case, the system can be modelled by a three dimensional Markov process $(X_t)$ with a state $(S,I,R),$ denoting the numbers of susceptible, infected, and removed, nodes, respectively.  Given $X_t=(S,I,R),$  the distribution of $X_{t+1}$ is determined by generating the number of newly infected nodes, which has the binomial distribution with parameters $S$ and $p=1-(1-\beta)^I$ (because each of the susceptible nodes is independently infected with probability $p.$)  The next cell contains a code for simulation of such a process.

```python
#  Simulation of Reed Frost model for fully connected graph (aka mean field model)
#  X[t,:]=[number susceptible, number infected, number removed] at time t
#  Since odeint wants t first in trajectories X[t,i], let's do that consistently

n=100       # Number of nodes
I_init=6    # Number of nodes initially infected, the others are initially susceptible

c=2.0     # Use a decimal point when specifying c.
beta=c/n  # Note that if n nodes each get infected with probability beta=c/n,
          # then c, on average, are infected.  

T=100  # maximum simulation time
X = np.zeros((T,3),dtype=np.int)


X[0,0], X[0,1] = n-I_init, I_init

t=0
while t<T and X[t,1]>0:   # continue (up to time T) as long as there exist infected nodes
    newI=np.random.binomial(X[t,0],1.0-(1.0-beta)**X[t,1])  # number of new infected nodes
    X[t+1,0]=X[t,0]-newI
    X[t+1,1]=newI
    X[t+1,2]=X[t,1]+X[t,2]
    t=t+1

print "beta=",beta    
print "state sequence X, one time step per row:\n", X[0:t+1,:]

plt.figure()
plt.plot(X[0:t,0], 'g')
plt.plot(X[0:t,1], 'r')
plt.plot(X[0:t,2], 'b')
plt.title('SIR Model for complete graph')
plt.xlabel('time')
plt.ylabel('number of nodes')
```

## Part 2: An Investment Problem

<h1>Analysis:</h1>

<p>From Lab2, we know how to standardize a random variable with mean = 0 and variance = 1:</p>

$E[X] = E\left[\frac{Y-\mu_y}{\sigma_y}\right] = \frac{1}{\sigma_y}E[Y-\mu_y] = \frac{1}{\sigma_y}(E[Y]-\mu_y) = 0$

$\mbox{Var}(X) = \mbox{Var}\left(\frac{Y-\mu_y}{\sigma_y}\right) = \frac{1}{\sigma_y^2}\mbox{Var}(Y-\mu_y) = \frac{\mbox{Var}[Y]}{\sigma_y^2} = 1$

So to standardize any random variable, we simply need to subtract the mean and divide by the standard deviation. This is useful because we often know or can find the CDF of a standard variable but not necessarily a non-standard random variable. For example, $Y$ is again a non-standard variable and $X$ is the standardized version of Y. We want to determine the probability that Y is less than some value c, but only have the CDF of X. We can determine the probability by the following:

$P\{Y \le c\} = P\{Y-\mu_y \le c-\mu_y\} = P\{\frac{Y-\mu_y}{\sigma_y} \le \frac{c-\mu_y}{\sigma_y}\}= P\{X \le \frac{c-\mu_y}{\sigma_y}\}$

Therefore, for this problem, I need to create two indeoendent uniformly distributed rv U and V, then standardize it with mean and variance

<h2>Math:</h2>

$E[S] = E[a + bU + cV] = a_S + b_S E[U] + c_S E[V] = 1.10$

$Var(S) = Var[a + bU + cV] = b^2_S Var[U] + c^2_S Var[V] = 0.15^2$

$E[B] = E[a + bU + cV] = a_B + b_B E[U] + c_B E[V] = 1.0$

$Var(B) = Var[a + bU + cV] = b^2_B Var[U] + c^2_B Var[V] = 0.15^2$

$E[T] = 1.02$

$Var(T) = 0$

$\rho_{S,B} = \frac{Cov(S,B)}{\sigma_S \sigma_B} = -0.8$


U and V are standardized random variable, thus $E[U] = E[V] = 0$, $Var[U] = Var[V] = 1$

$a_S = 1.1, a_B = 1.0$

$b^2_S  + c^2_S = 0.15^2$

$b^2_B  + c^2_B = 0.15^2$

$Cov(S,B) = \sigma_S \sigma_B \rho = -0.018 = E[(S - E[S])(B - E[B])] = E[(b_S U + c_S V)(b_B U + c_B V)] = E[b_s b_B U^2 + b_s c_B UV + c_S b_B VU + c_S c_B V^2]$

since $E[U] = E[V] = 0$, $Var[U] = Var[V] = 1$ 

thus $E[U^2] = Var[U] + E[U]^2 = Var[U],  E[V^2] = Var[V]$

$Cov(S,B) = b_s b_B  + c_S c_B  = -0.018 $

```python
##  Student response
from scipy.optimize import fsolve

N = 10000
res=[]
R=1
u = 0.5
var = 1/12. 
for i in range(N):
    #########
    U = np.random.uniform(0,1) #generate a uniform distributed random variable on [0,1]
    V = np.random.uniform(0,1)
    U = (U-u)*1./var**0.5 # Standardize
    V = (V-u)*1./var**0.5
    w = R*np.array([0.5,0.5,0])  ## distribute last year's return every year, so it is a inductive form
    #########
    
    ############### find bs, cs,bb, cb, suppose bs = 0.09, then cs = 0.12, use the remaining equations to find bb,cb
    bs = 0.09
    cs = 0.12
    
    
    def f(x):
        x0 = float(x[0])
        x1 = float(x[1])
        return [x[0]**2 + x[1]**2 - 0.15**2, 
                bs*x[0] + cs*x[1] + 0.018
                ]
    
    result = fsolve(f,[1,1])   ## use scipy.optimize.fsolve to find roots
    
    bb = result[0] #0.09*（0.0225-cb^2）^0.5 + 0.12*cb = -0.018
    cb = result[1]
    
    ###############
    
    #### generate S,B,T vector
    S = 1.10 + bs*U + cs*V # stock
    B = 1.0 + bb*U + cb*V # bond
    T = 1.02 #treasury bill
    Portfolio = np.array([S,B,T])
    ####
    
    #R represents the one year return for investing a fraction w0 in stocks, a fraction w1 in bonds, and a fraction w2 in T-bills. 
    R = np.dot(w,Portfolio)

    res.append(Portfolio)
res = np.asarray(res)

################################### Adapt Lab7 code to print out results
print 'The mean value of Stock is: \n', np.mean(res[:,0]),'\n'
print 'The mean value of Bond is: \n', np.mean(res[:,1]),'\n'
print 'The mean value of T-Bill is: \n', np.mean(res[:,2]),'\n'

print 'The standard deviation of Stock is: \n', np.std(res[:,0]),'\n'
print 'The standard deviation of Bond is: \n', np.std(res[:,1]),'\n'
print 'The standard deviation of T-Bill is: \n', np.std(res[:,2]),'\n'

print 'The final gain is: \n', R,'\n'
mean = np.mean(res, axis=0).reshape(1,-1)
print 'The empirical mean is: \n', mean,'\n'
CovarianceMatrix = np.dot((res - mean).T, res - mean)/res.shape[1]
print 'covariance matrix: \n', CovarianceMatrix,'\n'
variances=np.diag(CovarianceMatrix).reshape(-1,1)  # column vector of variances
std_dev=np.sqrt(variances)    # vector of stdev's

print "variances=  (diagonal entries of the covariance matrix)", '\n', \
        variances,'\n\n', "standard deviations=",'\n', std_dev, '\n'

correlation_matrix=CovarianceMatrix/std_dev/std_dev.T
      # In numpy, division of a matrix (a_ij) by a row vector (v_j) divide a_ij by v_j
      # and division of (a_ij) by a column vector (v_i) divides a_ij by v_i
print "correlation matrix=  (i.e. matrix of correlation coefficients)",'\n', correlation_matrix,'\n'
###################################


area = 30
plt.scatter(res[0:100,0],res[0:100,1], area, alpha = 0.5),
plt.title('Relationship between Stock and Bond')
plt.xlabel('Stock values')
plt.ylabel('Bond values')
plt.show()
```

$\widehat{E}\left[X_3 \bigg| \left(\begin{array}{c} X_1  \end{array}\right) \right]$

 the minimum mean square error linear estimator is given by
$$ \widehat{E}[Y | X] =L^*(X) = E[Y] +\frac{ \mbox{Cov}(Y,X) }{\mbox{Var}(X) } (X-E[X])$$
$$= E[Y] + \frac{\sigma_X \sigma_Y \rho_{X,Y}}{\sigma_X^2} \left(X-E[X]\right)$$
$$= E[Y] + \sigma_Y \rho_{X,Y} \left( \frac{X-E[X]}{\sigma_X } \right)$$
and the mean square error is given by:
$$(Y-\widehat{E}[Y | X])^2 = \mbox{Var}(Y) - \frac{ \mbox{Cov}(Y,X)^2}{ \mbox{Cov}(X,X)} = \sigma_Y^2(1-\rho_{X,Y}^2),$$   where $ \rho_{X,Y} $ is the correlation coefficient between $X$ and $Y.$

```python
# create a covariance matrix for X1, X2,X3
## X2 and X3 are uncorrolated, that means Cov(X2, X3) = 0
CovarianceMatrix = np.array([[7,-4,-2],[-4,9,0],[-2,0,3]]) ## [2][3] and [3][2] should be zero because X2 and X3 is uncorrelated
print CovarianceMatrix
eigs = np.linalg.eigvals(CovarianceMatrix)
print eigs
assert (eigs < 0.).all, 'not positive definite'
```

```python
# Calculate mean square error and root mean square error for predicting
# X3 by using any one of the first three scores
# Think of Y as representing X3

variance = np.diag(CovarianceMatrix).reshape(-1,1)
std = np.sqrt(variance)
CorrelationMatrix = CovarianceMatrix/std/std.T

CorYX=CorrelationMatrix[2,0:2]  # correlation coefficients between Y and others
VarY=CovarianceMatrix[2,2]
print "Mean square errors =", VarY *(1-CorYX**2)
print  "sqare roots of MSEs", np.sqrt(VarY*(1-CorYX**2))
```

```python
mean_x = np.random.rand(3) # generate random mean values of X1, X2 and X3
print mean_x
EX=mean_x[0:2]
EY=mean_x[2]
CovXX=CovarianceMatrix[0:2,0:2]
CovXY=CovarianceMatrix[0:2,2]
CovYX=CovarianceMatrix[2,0:2]

print CovYX
assert all(CovXY==CovYX.T), "CovXY is not the transpose of CovYX"

print "weight vector=", np.dot(CovXY,np.linalg.inv(CovXX))
print "the constant=", EY- np.dot( np.dot(CovXY,np.linalg.inv(CovXX)),EX)

MSE= VarY -np.dot(CovYX,np.dot(np.linalg.inv(CovXX),CovXY))
print "MSE for multiple regression", MSE, "sqare root of MSE",np.sqrt(MSE)
```

#### So the best linear estimator of X3 given X1,X2 is $$\widehat{E}\left[X_3 \bigg| \left(\begin{array}{c} X_1 \\ X_2 \end{array}\right) \right]= (-0.383)X1+(-0.17)X2 + 0.83 $$ and the sqare root of the MSE (when applied to the data) is about 1.4947, which is smaller than 1.558 ($\widehat{E}\left[X_3 \bigg| \left(\begin{array}{c} X_1  \end{array}\right) \right]$).  Therefore, there exists a matrix  $$Covariance Matrix =\left( \begin{array}{cc}  7 & -4 & 2 \\ -4 & 9 & 0 \\ -2 & 0 & 3  \end{array}\right)$$ such that X2,X3 is uncorrelated but X2 is useful for linear estimation of X3 when X1 is also used.
